```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom, norm
import pickle
import matplotlib.pyplot as plt
import torch as tc
```

```python
with open('../data/1215-2220/results.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
def prepare_boxplot(dat):
    dat = [d[0] for d in dat]
    dat = tc.tensor(dat)
    dat = tc.t(dat)
    return dat
```

```python
data = [d[0] for d in data[0][0]]
data = [d[0] for d in data]
```

```python
mean = [d.mean() for d in data]
sd = [d.std() for d in data]
```

```python
plt.boxplot(mean)
xmin, xmax = plt.xlim()
plt.hlines(7.25, xmin, xmax)
plt.xticks([])
plt.ylabel("Mean")
```

```python
plt.boxplot(sd)
xmin, xmax = plt.xlim()
plt.hlines(0.91, xmin, xmax, alpha=0.6)
plt.xticks([])
plt.ylabel("Standard Deviation")
plt.ylim([0.5, 1.5])
```
