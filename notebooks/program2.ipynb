```python
import pickle
import matplotlib.pyplot as plt
import torch as tc
import numpy as np
```

```python
with open('../data/plots_rej_0.pkl', 'rb') as f:
    p2r0 = pickle.load(f)
    
with open('../data/plots_rej_1.pkl', 'rb') as f:
    p2r1 = pickle.load(f)
    
with open('../data/p3/plots_rej_p3_0.pkl', 'rb') as f:
    p3r0 = pickle.load(f)
    
    
with open('../data/p3/plots_rej_p3_1.pkl', 'rb') as f:
    p3r1= pickle.load(f)
```

```python
def prepare_boxplot(dat):
    dat = [d[0] for d in dat]
    dat = tc.tensor(dat)
    dat = tc.t(dat)
    return dat
```

```python
p2r0_box = prepare_boxplot(p2r0)
p2r1_box = prepare_boxplot(p2r1)
p3r0_box = prepare_boxplot(p3r0)
p3r1_box = prepare_boxplot(p3r1)
```

```python
p3r1_box
```

```python
plt.boxplot(p2r0_box)
```

```python
ticks = ['1', '2']

def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)

plt.figure()

bpl = plt.boxplot(p2r0_box, positions=np.array(range(len(p2r0_box)))*2.0-0.6, sym='', widths=0.8)
bpr = plt.boxplot(p2r1_box, positions=np.array(range(len(p2r0_box)))*2.0+0.6, sym='', widths=0.8)
set_box_color(bpl, '#D7191C') # colors are from http://colorbrewer2.org/
set_box_color(bpr, '#2C7BB6')

# draw temporary red and blue lines and use them to create a legend
plt.plot([], c='#D7191C', label='SMC')
plt.plot([], c='#2C7BB6', label='Rejuvenated SMC')
plt.legend()

plt.xticks(range(0, len(ticks) * 2, 2), ticks)
plt.xlim(-2, len(ticks)*2)
plt.ylim(0, 0.7)
plt.tight_layout()
plt.show()
```
